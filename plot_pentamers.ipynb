# plot pentamar like + sign

```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from the file
filename = '2d_RSAD.txt'
x_data = []
y_data = []

with open(filename, "r") as file:
    for line in file:
        # Strip whitespace and split into two columns
        stripped_line = line.strip()
        if stripped_line:  # Ensure the line is not empty
            try:
                x, y = map(float, stripped_line.split())  # Split line into x and y
                x_data.append(x)
                y_data.append(y)
            except ValueError:
                print(f"Warning: Could not parse line '{stripped_line}'. Skipping...")

# Check if data was successfully read
if not x_data or not y_data:
    print("Error: No valid data found in the file.")
    exit()

# Plot the data with specified marker style
plt.plot(x_data, y_data, '+', mew=15, ms=45, alpha=0.7)
plt.plot(x_data, y_data, 'o', alpha=0.7)

# Customize plot
plt.title("Particles Before diffusion")
plt.xlabel("X Values")
plt.ylabel("Y Values")
# Set major grid lines at integer steps
plt.grid(True, which='major', linestyle='--', linewidth=0.7)
plt.xticks(np.arange(min(x_data), max(x_data) + 1, 1), [])  # Grid at every 1 unit, no labels
plt.yticks(np.arange(min(y_data), max(y_data) + 1, 1), [])  # Grid at every 1 unit, no labels


# Step 5: Show the plot
plt.show()
```
