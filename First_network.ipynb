```python
import networkx as nx
import matplotlib as plt
```

```python
# To create an empty undirected graph
G = nx.Graph()

G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(7)
G.add_node(9)

G.add_edge(1,2)
G.add_edge(3,1)
G.add_edge(2,4)
G.add_edge(4,1)
G.add_edge(9,1)
G.add_edge(1,7)
```

```python
nx.draw(G, with_labels=True)
```

```python
# To get all the nodes of a graph
node_list = G.nodes()
print("#1")
print(node_list)
```

```python
# To get all the edges of a graph
edge_list = G.edges()
print("#2")
print(edge_list)
```

```python
# To remove a node of a graph
G.remove_node(3)
node_list = G.nodes()
print("#3")
print(node_list)
```

```python
# To remove an edge of a graph
G.remove_edge(1,2)
edge_list = G.edges()
print("#4")
print(edge_list)
```

```python
# To find number of nodes
n = G.number_of_nodes()
print("#5")
print(n)
  
```

```python
nx.draw(G, with_labels='True')
# To find degree of a node
# d will store degree of node 2
d = G.degree(2)
print("#7")
print(d)
 
```

```python
G.add_node(5)
G.add_nodes_from(range(10,21))

G2 = nx.Graph()
G2.add_nodes_from(range(5))  # Example: add 5 nodes (0,1,2,3,4)

# Now add edges between all pairs
# for i in G2.nodes:
#     for j in G2.nodes:
#         if i != j:
#             G2.add_edge(i, j)

edges = []
for i in G2.nodes:
    for j in G2.nodes:
        if i < j:  # avoid duplicate edges in undirected graph
            edges.append((i, j))
G2.add_edges_from(edges)

nx.draw(G2)
```
